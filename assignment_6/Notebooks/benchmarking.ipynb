```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
fpath = "../data/" # All the timings are stored in the data folder!
```

### Get box plots for spatial query timings

```python
layer_data = []
with open(f'{fpath}SpatialTimingsLayer3Animal1Bregma021.csv', 'r') as file:
    csv_reader = csv.reader(file, delimiter='\t')
    next(csv_reader)
    for row in csv_reader:
        layer_data.append(float(row[0].split(":")[1]) + float(row[0].split(":")[2])/60)

assignment6_layer1_stats = [float(x)*60 for x in layer_data]
plt.boxplot(assignment6_layer1_stats);
plt.ylabel("Query time (s)");
plt.xticks([1], [''])
plt.title("Spatial query times for z-slice 3, animal 1, bregma 0.21")
```
